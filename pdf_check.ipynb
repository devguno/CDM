```python
import fitz  # PyMuPDF
import re

pdf_path = r"C:\extract\2109_16_47094686.pdf"
pdf_doc = fitz.open(pdf_path)
page = pdf_doc.load_page(0)
extracted_text = page.get_text("text")

lines = extracted_text.split("\n")
print("Extracted Text (Line by Line):")
for line in lines:
    print(line)

# 정보 저장용 변수 초기화
patient_name = None
patient_id = None
age = None
gender = None

# 줄 단위로 탐색하여 키워드 이전 줄의 값을 추출
for i in range(1, len(lines)):
    if "Patient Name" in lines[i]:
        patient_name = lines[i - 1].strip()
    elif "ID" == lines[i].strip():
        patient_id = lines[i - 1].strip()
    elif "Age" in lines[i]:
        age = lines[i - 1].strip().split()[0]  # "64 yr" 같은 경우 숫자만 추출
    elif "Gender" in lines[i]:
        gender = lines[i - 1].strip()

# 결과 출력
print("\n--- Extracted Info ---")
print("Patient Name:", patient_name if patient_name else "Not found")
print("ID:", patient_id if patient_id else "Not found")
print("Age:", age if age else "Not found")
print("Gender:", gender if gender else "Not found")
```
