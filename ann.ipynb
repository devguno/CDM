```python
import os
import shutil
from tqdm import tqdm
 
# 원래 폴더 이름 리스트
folders_to_copy = [
    '163', '164', '173', '1927', '213', '216', '202112', '167', '176', '188',
    '1941', '175', '219', '184', '178', '162', '202111', '202110', '1928', '217',
    '169', '172', '1978', '202106', '161', '166', '251', '181', '198', '250',
    '202107', '1923', '212', '158', '179', '1960', '218', '1963', '1908', '159',
    '168', '234', '1970', '171', '160', '202108', '174', '170', '201108', '202109',
    '177', '215'
]
 
# 숫자로 변환하여 오름차순 정렬
folders_to_copy = sorted(folders_to_copy, key=int)
 
source_dir = r"E:\holter_finish"
dest_dir = r"C:\child_ann"
 
# tqdm를 사용하여 진행 상황을 표시
for folder in tqdm(folders_to_copy, desc="폴더 복사 진행", unit="folder"):
    src_path = os.path.join(source_dir, folder)
    dest_path = os.path.join(dest_dir, folder)

    if os.path.exists(src_path) and os.path.isdir(src_path):
        try:
            # Python 3.8 이상에서는 dirs_exist_ok=True 옵션 사용 가능
            shutil.copytree(src_path, dest_path, dirs_exist_ok=True)
            print(f"\n'{src_path}'를 '{dest_path}'로 복사했습니다.")
        except Exception as e:
            print(f"\n'{src_path}' 복사 중 에러 발생: {e}")
    else:
        print(f"\n폴더 '{src_path}'가 존재하지 않습니다.")
 
```
